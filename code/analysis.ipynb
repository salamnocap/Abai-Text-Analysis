```python
import nltk
nltk.download('punkt')
```

```python
def read_and_tokenize(file_path: str):
    with open(file_path, 'r') as file:
        text = file.read()
    return nltk.word_tokenize(text)
```

# English

```python
en_part1 = read_and_tokenize('en/part1.txt')
en_part2 = read_and_tokenize('en/part2.txt')
```

```python
print('Count of words in part1:', len(en_part1))
print('Count of words in part2:', len(en_part2))
```

# Russian

```python
ru_part1 = read_and_tokenize('ru/kniga1.txt')
ru_part2 = read_and_tokenize('ru/kniga2.txt')
ru_part3 = read_and_tokenize('ru/kniga3.txt')
ru_part4 = read_and_tokenize('ru/kniga4.txt')
```

```python
print('Count of words in part1:', len(ru_part1))
print('Count of words in part2:', len(ru_part2))
print('Count of words in part3:', len(ru_part3))
print('Count of words in part4:', len(ru_part4))
```

# Kazakh

```python
kaz_part1 = read_and_tokenize('kaz/kitap1.txt')
kaz_part2 = read_and_tokenize('kaz/kitap2.txt')
kaz_part3 = read_and_tokenize('kaz/kitap3.txt')
kaz_part4 = read_and_tokenize('kaz/kitap4.txt')
```

```python
print('Count of words in part1:', len(kaz_part1))
print('Count of words in part2:', len(kaz_part2))
print('Count of words in part3:', len(kaz_part3))
print('Count of words in part4:', len(kaz_part4))
```
